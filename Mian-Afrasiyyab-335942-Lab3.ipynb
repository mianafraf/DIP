# Task 1: Rescaling the Image


```python
import matplotlib.pyplot as plt
```

```python
from skimage import data, color
from skimage.transform import rescale
```

```python
image = color.rgb2gray(data.astronaut())
```

```python
image_rescaled = rescale(image, 0.1, anti_aliasing=False)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")

ax[1].imshow(image_rescaled, cmap='gray')
ax[1].set_title("Rescaled 1/4")
```

# Task 2: Rescale an image using Resize function

```python
from skimage.transform import resize
```

```python
image = color.rgb2gray(data.astronaut())
```

```python
image_resized = resize(image, (100, 200),
                       anti_aliasing=True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")

ax[1].imshow(image_resized, cmap='gray')
ax[1].set_title("Resized (100,200)")
```

```python
print("Size of Actual Image: ", image.shape)
print("Size of Rescaled Image: ", image_rescaled.shape)
print("Size of Resized Image: ", image_resized.shape)
```

# Task 3: 1D interpolation


```python
from scipy import interpolate
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, num=10, endpoint=True)
y = [5,3,7,10,5,7,10,10,7,8]
```

```python
linear = interpolate.interp1d(x, y, kind='linear')
quadratic= interpolate.interp1d(x,y, kind='quadratic')
cubic = interpolate.interp1d(x,y, kind='cubic')
```

```python
x_new=np.linspace(0,10, num=100, endpoint=True)
```

```python
y_lin=linear(x_new)
y_quad=quadratic(x_new)
y_cub=cubic(x_new)
```

```python
plt.plot(x,y,'o',label='data')
plt.plot(x_new,y_lin,'-',label='linear')
plt.plot(x_new,y_quad,'--',label='quadratic')
plt.plot(x_new,y_cub,'--',label='cubic')
plt.legend()
plt.show()
```

# Task 4: 2D interpolation

```python
x=np.linspace(0,99,100)
```

```python
y=np.linspace(0,99,100)
```

```python
xx, yy =np.meshgrid(x,y)
```

```python
z = np.sin(xx**2+yy**2)
```

```python
f = interpolate.interp2d(x, y, z, kind='cubic')
```

```python
xnew = np.arange(-5.01, 5.01, 1e-2)
ynew = np.arange(-5.01, 5.01, 1e-2)
```

```python
znew = f(xnew, ynew)
```

```python
plt.plot(x, z[0, :], 'ro-', xnew, znew[0, :], 'b-')
plt.show()
```
